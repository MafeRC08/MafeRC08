````r
# ════════════════════════════════════════════════════════════
#  CELDA 5: ABUNDANCIA RELATIVA - CPUE
#  Fórmula: CPUE = N / (número de nasas × tiempo en horas)
#  Evento 1: 13 nasas × 5 días × 24 h = 1560 horas-trampa
#  Evento 2: 5 nasas × 3 días × 24 h = 360 horas-trampa
# ════════════════════════════════════════════════════════════
cat("\n══════════════════════════════════════════\n")
cat("5.4.2 ABUNDANCIA RELATIVA (CPUE)\n")
cat("══════════════════════════════════════════\n")

# Define evento1 and evento2 based on the date column 'fecha' from 'datos'
# Assuming 'datos' is the main dataset and 'fecha' column is available and correctly parsed.
# Based on comments, Evento 1 is Oct-Nov 2025, Evento 2 is Jan 2026.
# If all data in 'datos' is from Jan 2026, then 'evento1' will be empty.
evento1 <- datos %>% filter(fecha >= as.Date("2025-10-01") & fecha <= as.Date("2025-11-30"))
evento2 <- datos %>% filter(fecha >= as.Date("2026-01-01") & fecha <= as.Date("2026-01-31"))

# Parámetros de muestreo
nasas_e1    <- 13;  dias_e1 <- 4;  horas_e1 <- nasas_e1 * dias_e1 * 24
nasas_e2    <- 5;   dias_e2 <- 2;  horas_e2 <- nasas_e2 * dias_e2 * 24
horas_total <- horas_e1 + horas_e2

N_e1 <- nrow(evento1)
N_e2 <- nrow(evento2)
N_total <- nrow(datos)

CPUE_e1    <- N_e1    / horas_e1
CPUE_e2    <- N_e2    / horas_e2
CPUE_total <- N_total / horas_total

cat(sprintf("\nEvento 1 (oct-nov 2025):\n"))
cat(sprintf("  Nasas: %d | Días: %d | Esfuerzo: %d horas-trampa\n", nasas_e1, dias_e1, horas_e1))
cat(sprintf("  Capturas: %d individuos\n", N_e1))
cat(sprintf("  CPUE = %.4f individuos/hora-trampa\n", CPUE_e1))

cat(sprintf("\nEvento 2 (ene 2026):\n"))
cat(sprintf("  Nasas: %d | Días: %d | Esfuerzo: %d horas-trampa\n", nasas_e2, dias_e2, horas_e2))
cat(sprintf("  Capturas: %d individuos\n", N_e2))
cat(sprintf("  CPUE = %.4f individuos/hora-trampa\n", CPUE_e2))

cat(sprintf("\nTotal general:\n"))
cat(sprintf("  Esfuerzo total: %d horas-trampa\n", horas_total))
cat(sprintf("  Capturas totales: %d individuos\n", N_total))
cat(sprintf("  CPUE total = %.4f individuos/hora-trampa\n", CPUE_total))
````

````r
# ════════════════════════════════════════════════════════════
#  CELDA 6: ESTIMADOR DE CHAPMAN (1951)
#  N̂ = [(M+1)(C+1) / (R+1)] - 1
#  M = marcados en evento 1
#  C = capturados en evento 2
#  R = recapturados en evento 2 (individuos con marca del evento 1)
# ════════════════════════════════════════════════════════════
cat("\n══════════════════════════════════════════\n")
cat("5.4.2 ESTIMADOR POBLACIONAL DE CHAPMAN\n")
cat("══════════════════════════════════════════\n")

M <- nrow(evento1)   # marcados en evento 1
C <- nrow(evento2)   # capturados en evento 2

# Recapturas: individuos del evento 2 cuyos códigos aparecen en evento 1
codigos_e1 <- evento1$`general-codigo_marcaje`
codigos_recaptura_e2 <- datos %>%
  filter(fecha >= as.Date("2026-01-09")) %>% # Assuming this filter is correct for recaptures in event 2
  pull(`general-codigo_marcaje`)
R <- sum(codigos_recaptura_e2 %in% codigos_e1)

N_chapman <- ((M + 1) * (C + 1) / (R + 1)) - 1

# Varianza e intervalo de confianza 95% (fórmula Chapman)
var_N <- ((M + 1) * (C + 1) * (M - R) * (C - R)) / ((R + 1)^2 * (R + 2))
SE_N  <- sqrt(var_N)
IC_inf <- round(N_chapman - 1.96 * SE_N)
IC_sup <- round(N_chapman + 1.96 * SE_N)

cat(sprintf("\nParámetros:\n"))
cat(sprintf("  M (marcados evento 1):     %d\n", M))
cat(sprintf("  C (capturados evento 2):   %d\n", C))
cat(sprintf("  R (recapturados c/marca):  %d\n", R))
cat(sprintf("\nEstimado poblacional:\n"))
cat(sprintf("  N̂ (Chapman) = %.1f individuos\n", N_chapman))
cat(sprintf("  Error estándar = %.2f\n", SE_N))
cat(sprintf("  IC 95%% = [%d, %d] individuos\n", IC_inf, IC_sup))
````

````r
# ════════════════════════════════════════════════════════════
#  CELDA 7: PRUEBAS DE NORMALIDAD (Shapiro-Wilk)
# ════════════════════════════════════════════════════════════
cat("\n══════════════════════════════════════════\n")
cat("5.4.1 PRUEBAS DE NORMALIDAD (Shapiro-Wilk)\n")
cat("══════════════════════════════════════════\n")

variables_morfo <- list(
  "Longitud del Caparazón"        = datos$longitud_del_caparazon,
  "Ancho del Caparazón"           = datos$ancho_del_caparazon,
  "Altura del Caparazón"          = datos$altura_del_caparazon,
  "Longitud del Plastrón"         = datos$longitud_del_plastron,
  "Longitud del Plastrón Segunda" = datos$longitud_del_plastron_segunda,
  "Ancho del Plastrón"            = datos$ancho_del_plastron,
  "Largo 3ra Uña"                 = datos$largo_de_la_3ra_una,
  "Long. Preanal de la Cola"      = datos$longitud_preanal_de_la_cola,
  "Long. Posanal de la Cola"      = datos$longitud_posanal_de_la_cola,
  "Peso"                          = datos$peso
)

cat(sprintf("\n%-35s %8s %8s %s\n", "Variable", "W", "p-valor", "Distribución"))
cat(strrep("-", 65), "\n")

for (nombre in names(variables_morfo)) {
  v <- na.omit(variables_morfo[[nombre]])
  if (length(v) >= 3 & length(v) <= 5000) {
    sw <- shapiro.test(v)
    distrib <- ifelse(sw$p.value >= 0.05, "Normal ✓", "No normal ✗")
    cat(sprintf("%-35s %8.4f %8.4f %s\n", nombre, sw$statistic, sw$p.value, distrib))
  }
}
cat("\nNota: p \u2265 0.05 \u2192 distribución normal; p < 0.05 \u2192 no normal\n")
````

````r
# ════════════════════════════════════════════════════════════
#  CELDA 8: ESTADÍSTICAS DESCRIPTIVAS
# ════════════════════════════════════════════════════════════
cat("\n══════════════════════════════════════════\n")
cat("5.4.1 ESTADÍSTICAS DESCRIPTIVAS\n")
cat("══════════════════════════════════════════\n")

estadisticas <- function(variable, nombre) {
  cat("\n-----------------------------\n")
  cat("VARIABLE:", nombre, "\n")
  cat("-----------------------------\n")
  print(summary(variable))
  cat("Media:              ", round(mean(variable, na.rm = TRUE), 2), "\n")
  cat("Mediana:            ", round(median(variable, na.rm = TRUE), 2), "\n")
  cat("Desviación estándar:", round(sd(variable, na.rm = TRUE), 2), "\n")
  cat("Varianza:           ", round(var(variable, na.rm = TRUE), 2), "\n")
  cat("Rango:              ", paste(round(range(variable, na.rm = TRUE), 2), collapse = " - "), "\n")
  cat("Coef. de variación: ", round((sd(variable, na.rm = TRUE) /
                                       mean(variable, na.rm = TRUE)) * 100, 2), "%\n")
  cat("n válidos:          ", sum(!is.na(variable)), "\n")
}

estadisticas(datos$longitud_del_caparazon,       "Longitud del Caparazón (mm)")
estadisticas(datos$ancho_del_caparazon,           "Ancho del Caparazón (mm)")
estadisticas(datos$altura_del_caparazon,          "Altura del Caparazón (mm)")
estadisticas(datos$longitud_del_plastron,         "Longitud del Plastrón (mm)")
estadisticas(datos$longitud_del_plastron_segunda, "Longitud del Plastrón Segunda (mm)")
estadisticas(datos$ancho_del_plastron,            "Ancho del Plastrón (mm)")
estadisticas(datos$largo_de_la_3ra_una,           "Largo de la 3ra Uña (mm)")
estadisticas(datos$longitud_preanal_de_la_cola,   "Longitud Preanal de la Cola (mm)")
estadisticas(datos$longitud_posanal_de_la_cola,   "Longitud Posanal de la Cola (mm)")
estadisticas(datos$peso,                          "Peso (g)")
````

````r
# ════════════════════════════════════════════════════════════
#  CELDA 9: PROPORCIÓN DE SEXOS Y DIMORFISMO SEXUAL
# ════════════════════════════════════════════════════════════
cat("\n══════════════════════════════════════════\n")
cat("5.4.4 PROPORCIÓN DE SEXOS\n")
cat("══════════════════════════════════════════\n")

sexos <- datos %>%
  filter(sexo %in% c("hembra", "macho")) %>%
  count(sexo) %>%
  mutate(porcentaje = round(n / sum(n) * 100, 1))
print(sexos)

H <- sexos$n[sexos$sexo == "hembra"]
M_n <- sexos$n[sexos$sexo == "macho"]
cat(sprintf("\nProporción H/M = %.2f\n", H / M_n))
cat(sprintf("Ratio H:M = %.2f : 1\n", H / M_n))

# Test Chi-cuadrado (¿difiere de 1:1?)
chi2 <- chisq.test(c(H, M_n), p = c(0.5, 0.5))
cat(sprintf("\nTest Chi-cuadrado (H\u2080: ratio 1:1):\n"))
cat(sprintf("  \u03c7\u00b2 = %.3f, gl = %d, p = %.4f\n",
            chi2$statistic, chi2$parameter, chi2$p.value))
cat(ifelse(chi2$p.value < 0.05,
           "  \u2192 Diferencia significativa del ratio 1:1 (p < 0.05)\n",
           "  \u2192 Sin diferencia significativa del ratio 1:1 (p \u2265 0.05)\n"))

# \u2500\u2500 Índice de dimorfismo sexual (Lovich y Gibbons, 1992) \u2500\u2500\u2500\ncat("\n══════════════════════════════════════════\n")
cat("5.4.4 ÍNDICE DE DIMORFISMO SEXUAL\n")
cat("(Lovich y Gibbons, 1992; Hays y McBee, 2010)\n")
cat("Fórmula: (LP Hembras / LP Machos) - 1\n")
cat("══════════════════════════════════════════\n")

LP_H <- mean(datos$longitud_del_plastron[datos$sexo == "hembra"], na.rm = TRUE)
LP_M <- mean(datos$longitud_del_plastron[datos$sexo == "macho"],  na.rm = TRUE)
IDS  <- (LP_H / LP_M) - 1

cat(sprintf("\nLP promedio hembras: %.2f mm\n", LP_H))
cat(sprintf("LP promedio machos:  %.2f mm\n", LP_M))
cat(sprintf("Índice de dimorfismo sexual = %.4f\n", IDS))
cat(ifelse(IDS > 0,
           "\u2192 Valor positivo: hembras de mayor tamaño que machos \u2713\n",
           "\u2192 Valor negativo: machos de mayor tamaño que hembras\n"))

# Gráfico proporción de sexos
ggplot(sexos, aes(x = sexo, y = n, fill = sexo)) +
  geom_bar(stat = "identity", width = 0.6, color = "white") +
  geom_text(aes(label = paste0(n, "\n(", porcentaje, "%)")),
            vjust = -0.4, size = 4.5, fontface = "bold") +
  scale_fill_manual(values = c("hembra" = "#E07B8A", "macho" = "#5B8DB8")) +
  labs(title = "Proporción de sexos (capturas nuevas)",
       subtitle = paste0("Ratio H:M = ", round(H/M_n, 2), ":1"),
       x = "Sexo", y = "Número de individuos") +
  theme_classic(base_size = 13) +
  theme(legend.position = "none") +
  ylim(0, max(sexos$n) * 1.25)
````

````r
# ── CLASES DE TALLA - FÓRMULA DE STURGES ──────────────────
cat("\n══════════════════════════════════════════\n")
cat("5.4.4 CLASES DE TALLA - FÓRMULA DE STURGES\n")
cat("══════════════════════════════════════════\n")

LC_vals   <- na.omit(datos$longitud_del_caparazon)
PESO_vals <- na.omit(datos$peso)

sturges_clases <- function(x, nombre_var) {
  N         <- length(x)
  R         <- max(x) - min(x)
  C_num     <- 1 + 3.322 * log10(N)
  n_clases  <- ceiling(C_num)
  amplitud  <- R / n_clases   # usar n_clases entero, no decimal

  cat(sprintf("\n%s:\n", nombre_var))
  cat(sprintf("  N = %d | Rango R = %.2f | Clases k = %.2f ≈ %d | Amplitud = %.2f\n",
              N, R, C_num, n_clases, amplitud))

  cortes <- seq(min(x), min(x) + amplitud * n_clases, by = amplitud)
  cortes[length(cortes)] <- max(x) + 0.01  # asegurar que el máximo quede incluido

  frec  <- cut(x, breaks = cortes, include.lowest = TRUE, right = FALSE)
  tabla <- as.data.frame(table(frec))
  tabla$porcentaje <- round(tabla$Freq / sum(tabla$Freq) * 100, 1)
  colnames(tabla)  <- c("Intervalo", "fi (frecuencia absoluta)", "% ")
  print(tabla)
  return(list(cortes = cortes, n_clases = n_clases, amplitud = amplitud))
}

res_LC   <- sturges_clases(LC_vals,   "Longitud del Caparazón (mm)")
res_PESO <- sturges_clases(PESO_vals, "Peso (g)")

# Histograma LC por sexo
datos %>%
  filter(!is.na(longitud_del_caparazon), sexo %in% c("hembra","macho")) %>%
  ggplot(aes(x = longitud_del_caparazon, fill = sexo)) +
  geom_histogram(breaks = res_LC$cortes, color = "white", alpha = 0.85) +
  scale_fill_manual(values = c("hembra" = "#E07B8A", "macho" = "#5B8DB8")) +
  labs(title = "Distribución de frecuencias - Longitud del Caparazón",
       subtitle = paste0("Intervalos por fórmula de Sturges (k = ", res_LC$n_clases, ")"),
       x = "Longitud del Caparazón (mm)", y = "Frecuencia absoluta (fi)", fill = "Sexo") +
  theme_classic(base_size = 13) +
  facet_wrap(~sexo, ncol = 1)

# Histograma Peso por sexo
datos %>%
  filter(!is.na(peso), sexo %in% c("hembra","macho")) %>%
  ggplot(aes(x = peso, fill = sexo)) +
  geom_histogram(breaks = res_PESO$cortes, color = "white", alpha = 0.85) +
  scale_fill_manual(values = c("hembra" = "#E07B8A", "macho" = "#5B8DB8")) +
  labs(title = "Distribución de frecuencias - Peso",
       subtitle = paste0("Intervalos por fórmula de Sturges (k = ", res_PESO$n_clases, ")"),
       x = "Peso (g)", y = "Frecuencia absoluta (fi)", fill = "Sexo") +
  theme_classic(base_size = 13) +
  facet_wrap(~sexo, ncol = 1)
````

````r
# ── CORRELACIÓN DE PEARSON ─────────────────────────────────
if (!require("corrplot")) install.packages("corrplot")
library(corrplot)

cat("\n══════════════════════════════════════════\n")
cat("5.4.3 CORRELACIÓN DE PEARSON\n")
cat("══════════════════════════════════════════\n")

# LC vs Peso
cor_data <- datos %>%
  filter(!is.na(longitud_del_caparazon), !is.na(peso))
r_test <- cor.test(cor_data$longitud_del_caparazon, cor_data$peso, method = "pearson")

cat(sprintf("\nCorrelación LC vs Peso:\n"))
cat(sprintf("  r = %.4f | t = %.3f | gl = %d | p = %.6f\n",
            r_test$estimate, r_test$statistic,
            r_test$parameter, r_test$p.value))
cat(ifelse(r_test$p.value < 0.05,
           "  → Correlación significativa (p < 0.05)\n",
           "  → No significativa (p >= 0.05)\n"))

# Gráfico dispersión
cor_data %>%
  filter(sexo %in% c("hembra","macho")) %>%
  ggplot(aes(x = longitud_del_caparazon, y = peso, color = sexo)) +
  geom_point(size = 3, alpha = 0.8) +
  geom_smooth(method = "lm", se = TRUE, alpha = 0.2) +
  scale_color_manual(values = c("hembra" = "#E07B8A", "macho" = "#5B8DB8")) +
  labs(title = "Correlación Longitud del Caparazón vs Peso",
       subtitle = paste0("r = ", round(r_test$estimate, 4),
                         " | p = ", round(r_test$p.value, 6)),
       x = "Longitud del Caparazón (mm)", y = "Peso (g)", color = "Sexo") +
  theme_classic(base_size = 13)

# Matriz de correlaciones
vars_cor <- datos %>%
  select(longitud_del_caparazon, ancho_del_caparazon, altura_del_caparazon,
         longitud_del_plastron, ancho_del_plastron, peso) %>%
  na.omit()

mat_cor <- cor(vars_cor, method = "pearson")
colnames(mat_cor) <- rownames(mat_cor) <- c("LC","AC","HC","LP","AP","Peso")

cat("\nMatriz de correlaciones:\n")
print(round(mat_cor, 3))

corrplot(mat_cor,
         method   = "color",
         type     = "upper",
         addCoef.col = "black",
         number.cex  = 0.8,
         tl.col   = "black",
         tl.srt   = 45,
         col      = colorRampPalette(c("#5B8DB8","white","#E07B8A"))(200),
         title    = "Matriz de correlación - Variables morfométricas",
         mar      = c(0, 0, 2, 0))
````

````r
# ════════════════════════════════════════════════════════════
#  CELDA 12: ÍNDICES MORFOMÉTRICOS
#  (Espinal, 1989; Cabrera, 2023)
#  LC/AC | LC/HC | LP/AP | LP/HC | AC/AP
# ════════════════════════════════════════════════════════════
cat("\n══════════════════════════════════════════\n")
cat("5.4.3 ÍNDICES MORFOMÉTRICOS POR SEXO\n")
cat("(Espinal, 1989; Cabrera, 2023)\n")
cat("══════════════════════════════════════════\n")

indices <- datos %>%
  filter(sexo %in% c("hembra", "macho")) %>%
  mutate(
    LC_AC = longitud_del_caparazon / ancho_del_caparazon,
    LC_HC = longitud_del_caparazon / altura_del_caparazon,
    LP_AP = longitud_del_plastron  / ancho_del_plastron,
    LP_HC = longitud_del_plastron  / altura_del_caparazon,
    AC_AP = ancho_del_caparazon    / ancho_del_plastron
  ) %>%
  group_by(sexo) %>%
  summarise(
    `LC/AC (prom ± DE)` = paste0(round(mean(LC_AC, na.rm=T),3),
                                  " ± ", round(sd(LC_AC, na.rm=T),3)),
    `LC/HC (prom ± DE)` = paste0(round(mean(LC_HC, na.rm=T),3),
                                  " ± ", round(sd(LC_HC, na.rm=T),3)),
    `LP/AP (prom ± DE)` = paste0(round(mean(LP_AP, na.rm=T),3),
                                  " ± ", round(sd(LP_AP, na.rm=T),3)),
    `LP/HC (prom ± DE)` = paste0(round(mean(LP_HC, na.rm=T),3),
                                  " ± ", round(sd(LP_HC, na.rm=T),3)),
    `AC/AP (prom ± DE)` = paste0(round(mean(AC_AP, na.rm=T),3),
                                  " ± ", round(sd(AC_AP, na.rm=T),3)),
    .groups = "drop"
  )

print(as.data.frame(indices))
````
